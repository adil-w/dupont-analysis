## **Dupont Formula and Analysis**
 
The Dupont Analysis relies on the fundamental ideas of the ROA calculation, which is an indicator of a company's ability to generate profits from its assets. The ROA formula is based on two building blocks: the asset turnover ratio (ATR) and the profit margin ratio (PMR). <br> - The ATR is the effectiveness and efficiency in generating revenue regardless of the profit generated. <br> - The PMR is the comparison between generating sales and efficiently incurring all expenses. <br> The dissection of the ROA formula is the essential foundation of the Dupont Analysis as it looks to provide insights into how a company generates its ROA. Below is the breakdown of the ROA formula: 

---

**Return on Assets (ROA)** = (Revenues/Assets) x (Net Income/Revenues)<br>
**Profit Margin Ratio (PMR)** = Net Income/Revenues <br>
**Asset Turnover Ratio (ATR)** = Revenues/Assets 

The Dupont Analysis helps assess management's performance in successfully optimizing the assets into positive outcomes for the business. This analysis focuses on two main management methods for maximizing the ROA: focus on generating *higher volume of sales* that impacts the asset turnover, or focus on *regulating expenses and selling price* that directly relates to the profit margin.

# Analysis of Changes in ROA <br> 
##### *Changes in ROA from 2016 to 2017 to 2018 for each of the 5 companies (**note Kroger does not have 2018 ratios**)*

**Delta** <br>
Delta maintained a steady ATR from 2016 to 2018 of around 0.7 to 0.8, whereas the PMR and ROA trends during the timeframe are relatively identical. The ROA and PMR decreased from 2016 to 2017 and then remained constant between 2017 and 2018. The drop in the ROA from 2016 to 2017 can be explained by the decrease in net income from a possible difficulty to tightly regulate selling price and expenses. Delta's managerial performance decreased from 2016 to 2017, which is about balancing between generating sales and efficiency of incurring all expenses. 

**Kroger** <br>
Kroger experienced a slight decrease in ROA and PMR from 2016 to 2017, while the ATR shot up in a year. Given the nature of the retail industry, Kroger was able to increase sales without necessarily increasing profit. While Kroger's management strategy may have been a focus on pushing sales, analyzing the financial statement will give a clearer picture of the strategies involved in the changes that occurred. 

**Starbucks** <br>
The ROA slightly increased between 2016 and 2017, but then decreased at a faster rate in 2018 then the rate at which it increased between 2016 and 2017. PMR slightly decreased from 2016 to 2017, whereas the ATR slightly increased. However, the ATR and PMR both changed at a much faster rate between 2017 and 2018 with ATR decreasing and PMR increasing. ATR and the ROA declined sharply between 2017 and 2018 that may be a result of the difficulty to effectively generate revenue from each dollar of Starbucks assets. Starbucks was able to increase PMR in 2018 due to its maturity in the industry and applying strategies to efficiently save costs.  

**Microsoft** <br>
While it is generally categorized as a software company, Microsoft is one of the pioneers of showing the breadth of software integration with other industries. The ROA has been on a steady decline since 2010 with slight increases in 2016 and 2017, but it has overall been on a decline. The company has assets in so many new and innovative sectors that it is difficult to manage sales and expenses efficiently and effectively until the departments reach maturity in the sub-sectors. That is clearly explained by the spikes and drops in PMR from year to year with a big jump from 2016 to 2017 and then a steep decline from 2017 to 2018. The ATR has been on a slow and steady decline since 2010, but note that it slightly increased from 2017 to 2018. 

**Amazon** <br>
The world's biggest e-Commerce company has continued to increase ROA since 2014 even though there was a slight decrease between 2016 and 2017 before a steep climb up in 2018. The PMR has also increased in 2018 with no change between 2016 and 2017. This indicates Amazon's impressive ability to effectively maintain sales while also managing all expenses. However, the ATR did decline from 2016 to 2017 with small increases between 2017 and 2018, which may be due to difficulty to maintain the previous year's level of revenue generation from each unit of asset.
